```python
# 6.step plot
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,20,4)
y=np.arange(0,10,2)
plt.step(x,y)
plt.show()
```

```python
# 7.2D histogram plot
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,20,4)
y=np.arange(0,10,2)
plt.hist2d(x,y)
plt.show()
```

```python
# 8.using plot with x-label
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,20,4)
y=np.arange(0,10,2)
plt.xlabel("x-value")
plt.bar(x,y)
plt.show()
```

```python
# 9.using plot with y-label
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,20,4)
y=np.arange(0,10,2)
plt.ylabel("y-value")
plt.bar(x,y)
plt.show()
```

```python
# 10.using plot with title
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,20,4)
y=np.arange(0,10,2)
plt.xlabel("x-value")
plt.ylabel("y-value")
plt.title("value prediction")
plt.bar(x,y)
plt.show()
```
