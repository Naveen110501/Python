```python
# 1.line plot
from matplotlib import pyplot as plt
x=[1,2,3,4,5]
y=[10,30,20,40,50]
plt.plot(x,y)
plt.show()
```

```python
# 2.bar plot
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,20,4)
y=np.arange(0,10,2)
plt.bar(x,y)
plt.show()
```

```python
# 3.histogram
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,20,4)
plt.hist(x)
plt.show()
```

```python
# 4.scatter plot
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,40,4)
y=np.arange(0,10)
plt.scatter(x,y)
plt.show()
```

```python
# 5.another way line plot
from matplotlib import pyplot as plt
x=np.arange(0,20,4)
y=np.arange(0,10,2)
plt.plot(x,y)
plt.show()
```
